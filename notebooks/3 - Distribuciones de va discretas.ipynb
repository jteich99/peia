## Distribuciones de variables aleatorias discretas

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson
import seaborn as sns

sns.set(style="whitegrid")
```

```python
def plot_poisson(lam=3):
    fig, axs = plt.subplots(1,2, figsize=(12,4))

    x_pois = np.arange(0, max(15, lam*3))
    pmf_pois = poisson.pmf(x_pois, lam)
    axs[0].bar(x_pois, pmf_pois, alpha=0.6, color='lightgreen', label='PMF')
    axs[0].set_title(f"Poisson (lambda={lam})")
    axs[0].set_xlabel("Número de eventos")
    axs[0].set_ylabel("Probabilidad")
    axs[0].set_yticks(np.arange(0, 0.5, 0.05))
    axs[0].legend()
    
    cdf_pois = poisson.cdf(x_pois, lam)
    axs[1].step(x_pois, cdf_pois, where='post', color='red', label='CDF')
    axs[1].set_title(f"Poisson (lambda={lam})")
    axs[1].set_xlabel("Número de eventos")
    axs[1].set_ylabel("Probabilidad acumulada")
    axs[1].legend()
    
    plt.tight_layout()
    plt.show()
```

```python
def plot_binomial(n=10, p_binom=0.5):
    fig, axs = plt.subplots(1,2, figsize=(12,4))

    x_binom = np.arange(0, n+1)
    pmf_binom = binom.pmf(x_binom, n, p_binom)
    cdf_binom = binom.cdf(x_binom, n, p_binom)
    axs[0].bar(x_binom, pmf_binom, alpha=0.6, color='skyblue', label='PMF')
    axs[0].set_title(f"Binomial (n={n}, p={p_binom})")
    axs[0].set_xlabel("Número de éxitos")
    axs[0].set_ylabel("Probabilidad")
    axs[0].legend()
    
    x_binom = np.arange(0, n+1)
    pmf_binom = binom.pmf(x_binom, n, p_binom)
    cdf_binom = binom.cdf(x_binom, n, p_binom)
    axs[1].step(x_binom, cdf_binom, where='post', color='red', label='CDF')
    axs[1].set_title(f"Binomial (n={n}, p={p_binom})")
    axs[1].set_xlabel("Número de éxitos")
    axs[1].set_ylabel("Probabilidad acumulada")
    axs[1].legend()

    plt.tight_layout()
    plt.show()
```

```python
plot_binomial(p_binom=0.15)
```

```python
plot_binomial(p_binom=0.5)
```

```python
plot_binomial(p_binom=0.9)
```

```python
plot_poisson(1)
```

```python
plot_poisson(4)
```

```python
plot_poisson(20)
```
